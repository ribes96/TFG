# Comparación del accuracy de varios modelos usando RBFSampler

**Este notebook tiene dos secciones. Las dos hacen lo mismo, pero la primera sección no normalizará los datos, y la segunda sí que lo hará. La normalización que haremos será la que usan en los ejemplos de scikit-learn**

## Sección 1: comparación de accuracy sin normalizar los datos

```python
from sklearn.datasets import load_digits
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.kernel_approximation import RBFSampler
```

```python
import numpy as np
import math
```

```python
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
```

```python
data = digits.data
target = digits.target
```

```python
N = data.shape[0]
prop_train = 2 / 3
N_train = math.ceil(N * prop_train)
N_test = N - N_train
```

**new_indices** retorna un array con los índices mezclados

```python
def new_indices():
    ind = np.arange(N)
    ind = np.random.choice(ind, len(ind), replace = False)
    return ind
```

```python
features = 50 * np.arange(50) + 50 # 50 100 150 .. 2500
scores_DT = []
scores_LinearSVC = []
```

Pasos dentro del bucle:
- Mezclar los datos
- Sacar features de los datos
- Separar en train y test
- Entrenar cada uno de los modelos
- Sacar los scores de cada uno de los modelos

```python
for feat in features:
    #Sacar los índices de la nueva ordenación
    indices = new_indices()
    train_indices = indices[:N_train]
    test_indices = indices[N_train:]

    # Reordenar datos y target (con la misma ordenación)
    dat = np.take(data, indices, axis = 0)
    targ = np.take(target, indices, axis = 0)
    
    # Crear y ajustar el sampler
    sampler = RBFSampler(n_components = feat)
    sampler.fit(dat)
    
    # Cambiar los datos por los features de los datos
    # (No es necesario hacer nada con los targets)
    dat = sampler.transform(dat)

    # Separar los features obtenidos en train y test
    dat_train = np.take(dat, train_indices, axis = 0)
    dat_test = np.take(dat, test_indices, axis = 0)

    targ_train = np.take(targ, train_indices, axis = 0)
    targ_test = np.take(targ, test_indices, axis = 0)
    
    # Crear los clasificadores
    dt_clf = DecisionTreeClassifier()
    svc_clf = LinearSVC()
    
    # Entrenar los clasificadores
    dt_clf.fit(dat_train, targ_train)
    svc_clf.fit(dat_train, targ_train)
    
    # Sacar los scores
    dt_score = dt_clf.score(dat_test, targ_test)
    svc_score = svc_clf.score(dat_test, targ_test)
    
    # Poner los scores en las listas de scores
    scores_DT.append(dt_score)
    scores_LinearSVC.append(svc_score)
```

```python
plt.plot(features, scores_DT, label = 'DecisionTree')
plt.plot(features, scores_LinearSVC, label = 'Linear SVC')
plt.title("Accuracy obtenida a medida que se incrementan la cantidad de features sin normalizar los datos")
plt.legend()
```
