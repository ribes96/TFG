raro, y el PCA tiene problemas con él. Voy a hacer aquí las pruebas con ello

```python
def find_error():
    data_train = get_data('mnist', n_ins = 2000)['data_train']
    pca = PCA(n_components=0.9, svd_solver="full")
    i = 0
    while True:
        print('{:<5}'.format(i), end = '')
        i += 1
        trans_data = resample(data_train)
        try:
            pca.fit(trans_data)
        except LinAlgError:
            return trans_data
```
